```
from sklearn.model_selection import train_test_split

predictors = train.drop(['Survived', 'PassengerId'], axis=1)
target = train["Survived"]
x_train, x_val, y_train, y_val = train_test_split(predictors, target, test_size = 0.22, random_state = 0)
```

```
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
gaussian = GaussianNB()
gaussian.fit(x_train, y_train)
y_pred = gaussian.predict(x_val)
precision_gaussian = precision_score(y_pred, y_val)
print(precision_gaussian)
recall_gaussian = recall_score (y_pred, y_val)
print(recall_gaussian)
f1_gaussian = f1_score(y_pred, y_val)
print(f1_gaussian)
acc_gaussian = round(accuracy_score(y_pred, y_val) * 100, 2)
print(acc_gaussian)
```

```
# Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_val)
precision_logreg = precision_score(y_pred, y_val)
print(precision_logreg)
recall_logreg = recall_score (y_pred, y_val)
print(recall_logreg)
f1_logreg = f1_score(y_pred, y_val)
print(f1_logreg)
acc_logreg = round(accuracy_score(y_pred, y_val) * 100, 2)
print(acc_logreg)
```
